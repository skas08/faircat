```python
import numpy as np
from collections import defaultdict

def gini(array):
    """Compute the Gini coefficient of an array"""
    array = np.sort(np.array(array))
    n = len(array)
    cumulative = np.cumsum(array)
    return (n + 1 - 2 * np.sum(cumulative) / cumulative[-1]) / n if n > 0 and cumulative[-1] > 0 else 0

def summarize_group_stats(group_dist):
    degree_list = []
    for deg, count in group_dist.items():
        degree_list.extend([float(deg)] * int(count))
    
    array = np.array(degree_list)
    stats = {
        "num_nodes": len(array),
        "total_degree": int(array.sum()),
        "min": np.min(array),
        "max": np.max(array),
        "mean": np.mean(array),
        "median": np.median(array),
        "std": np.std(array),
        "gini": gini(array)
    }
    return stats

# Replace your float keys if needed
group_0 = grouped_deg_dist[0]
group_1 = grouped_deg_dist[1]

stats_0 = summarize_group_stats(group_0)
stats_1 = summarize_group_stats(group_1)

print("Statistics for Attribute = 0:")
for k, v in stats_0.items():
    print(f"  {k}: {v:.3f}")

print("\nStatistics for Attribute = 1:")
for k, v in stats_1.items():
    print(f"  {k}: {v:.3f}")
```

```python
# distribution
import numpy as np
from collections import Counter

# Step 1: Compute node degrees from the sparse adjacency matrix
degrees = np.array(S.sum(axis=1)).flatten()

# Step 2: Split degrees by attribute group
deg_group_0 = degrees[attribute == 0]
deg_group_1 = degrees[attribute == 1]

# Step 3: Count frequency of each degree using Counter
hist_0 = Counter(deg_group_0)
hist_1 = Counter(deg_group_1)

# Step 4: Print or sort nicely
print("Degree distribution for attribute = 0:")
for deg, count in sorted(hist_0.items(), reverse=True):
    print(f"Degree {int(deg)}: {count} node(s)")

print("\nDegree distribution for attribute = 1:")
for deg, count in sorted(hist_1.items(), reverse=True):
    print(f"Degree {int(deg)}: {count} node(s)")
```

```python
import matplotlib.pyplot as plt

def plot_degree_histogram(deg_group, label, color):
    counts = Counter(deg_group)
    degrees, freqs = zip(*sorted(counts.items()))
    plt.plot(degrees, freqs, label=label, color=color)

plot_degree_histogram(deg_group_0, "Attribute = 0", "blue")
plot_degree_histogram(deg_group_1, "Attribute = 1", "red")


plt.xlabel("Degree")
plt.ylabel("Number of Nodes")
plt.title("Degree Distribution per Attribute Group")
plt.legend()
plt.grid(True)
plt.show()
```
